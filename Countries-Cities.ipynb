```python
from pyspark.sql import SparkSession

spark = (SparkSession.builder
.master("local[*]")
.appName('PySpark_Tutorial')
.getOrCreate()
)
```

```python
countries_csv_file = 'world_population.csv'
df_countries = spark.read.option('header', True).csv(countries_csv_file)
```

```python
df_countries.show(10, vertical=True)
```
